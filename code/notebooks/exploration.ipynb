```python
weights = {}
for i in range(len(df.iloc[:,3:].corr()['FinalGrade'])):
    if abs(df.iloc[:,3:].corr()['FinalGrade'].iloc[i]) > 0.14:
        print(df.iloc[:,3:].corr()['FinalGrade'].index[i], df.iloc[:,3:].corr()['FinalGrade'].iloc[i])
        weights[df.iloc[:,3:].corr()['FinalGrade'].index[i]]= df.iloc[:,3:].corr()['FinalGrade'].iloc[i]
    
```

```python
list(weights)[0] , list(weights.values())[0]
```

```python
improvability_score_list = []
for student in df[weights.keys()].values:
    improvability_score = 0
    for val in range(len(student)):
        if list(weights.values())[val] < 0 :
            improvability_score +=  student[val] / np.max(df[list(weights)[val]])
        else:
            improvability_score +=  ((-1) *student[val]) / np.max(df[list(weights)[val]])
    print(improvability_score)
    improvability_score_list.append(improvability_score)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs



# Calculate the Within-Cluster-Sum-of-Squares (WCSS) for different values of k
wcss = []
for k in range(1, 20):
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(df.iloc[:,3:])
    wcss.append(kmeans.inertia_)  # Inertia is the WCSS for that model

# Plot the elbow method curve
plt.figure(figsize=(8, 6))
plt.plot(range(1, 20), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.grid()
plt.show()
```

```python
optimal_k = 6  # Replace with the value you determined
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', max_iter=300, n_init=10, random_state=0)
kmeans.fit(df.iloc[:,3:])
cluster_assignments = kmeans.predict(df.iloc[:,3:])
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
from sklearn.decomposition import PCA

# Perform PCA to reduce data to 2 dimensions
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(df.iloc[:,3:])
```

```python
# Create a DataFrame for the reduced data
df = pd.DataFrame({'Dimension 1': reduced_data[:, 0], 'Dimension 2': reduced_data[:, 1], 'Cluster': cluster_assignments})

# Scatter plot with labels
sns.scatterplot(x='Dimension 1', y='Dimension 2', hue='Cluster', data=df, palette='viridis')
plt.title('KMeans Clustering')
plt.show()
```
